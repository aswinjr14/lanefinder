```python
def calibrateCamera(nx, ny):
    objp = np.zeros((nx*ny,3), np.float32)
    objp[:,:2] = np.mgrid[0:nx,0:ny].T.reshape(-1,2)

    # Arrays to store object points and image points from all the images.
    objpoints = [] # 3d points in real world space
    imgpoints = [] # 2d points in image plane.

    
    # Make a list of calibration images
    images = glob.glob('camera_cal/calibration*.jpg')
    image_no = 0
    # Step through the list and search for chessboard corners
    for fname in images:
        image_no += 1
        img = cv2.imread(fname)
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        #print(gray.shape[::-1])
        # Find the chessboard corners
        ret, corners = cv2.findChessboardCorners(gray, (nx,ny),None)

        # If found, add object points, image points
        if ret == True:
            objpoints.append(objp)
            imgpoints.append(corners)

            # Draw and display the corners
            img = cv2.drawChessboardCorners(img, (nx,ny), corners, ret)            
            mpimg.imsave("output_images/cameracalibration/camcal_" + str(image_no) + ".jpg", img)
            #cv2.imshow('img',img)
            #cv2.waitKey(500)
        #cv2.destroyAllWindows()
        
    ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1],None,None)
    return mtx, dist

def undistortImage(img, mtx, dist):
    # Use the OpenCV undistort() function to remove distortion
    undist = cv2.undistort(img, mtx, dist, None, mtx)    
    return undist

def gaussianBlur(img, kernel = 3):
    blurred = cv2.GaussianBlur(img, (kernel, kernel), 0)    
    return blurred

def combinedBinary(img,sobel_thresh_min = 20, sobel_thresh_max = 100, color_thresh_min = 170, color_thresh_max=255):
    # Convert to HLS color space and separate the S channel
    # Note: img is the undistorted image
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    s_channel = hls[:,:,2]

    # Grayscale image
    # NOTE: we already saw that standard grayscaling lost color information for the lane lines
    # Explore gradients in other colors spaces / color channels to see what might work better
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # Sobel x
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0) # Take the derivative in x
    abs_sobelx = np.absolute(sobelx) # Absolute x derivative to accentuate lines away from horizontal
    scaled_sobel = np.uint8(255*abs_sobelx/np.max(abs_sobelx))

    # Threshold x gradient
    sxbinary = np.zeros_like(scaled_sobel)
    sxbinary[(scaled_sobel >= sobel_thresh_min) & (scaled_sobel <= sobel_thresh_max)] = 1

    # Threshold color channel
    s_binary = np.zeros_like(s_channel)
    s_binary[(s_channel >= color_thresh_min) & (s_channel <= color_thresh_max)] = 1

    # Stack each channel to view their individual contributions in green and blue respectively
    # This returns a stack of the two binary images, whose components you can see as different colors
    color_binary = np.dstack(( np.zeros_like(sxbinary), sxbinary, s_binary)) * 255

    # Combine the two binary thresholds
    combined_binary = np.zeros_like(sxbinary)
    combined_binary[(s_binary == 1) | (sxbinary == 1)] = 1
    return color_binary, combined_binary
    
def roi_vertices(img, hor_uplimit, hor_downlimit):
    y_size, x_size = img.shape[:2]
    #setting the lowest point to 95% of the image size
    ver_down = 0.95
    #setting the top point to 65% of the image size
    ver_up = 0.65
    top_left = hor_uplimit
    top_right = 1 - hor_uplimit
    bottom_left = hor_downlimit
    bottom_right = 1 - hor_downlimit
    
    #4 points
    bottom_left_pt = [x_size * bottom_left, y_size * ver_down]
    bottom_right_pt=[x_size * bottom_right, y_size * ver_down]
    top_left_pt = [x_size * top_left, y_size * ver_up]
    top_right_pt = [x_size * top_right, y_size * ver_up]
    
    vertices = np.array([top_left_pt,top_right_pt,bottom_right_pt,bottom_left_pt], dtype=np.float32)
    return vertices

def unwrapImage(img):
    offset_l = 0
    offset_r = 1 - offset_l
    offset = 100
    width = img.shape[0]
    height = img.shape[1]
    src = roi_vertices(img,0.44,0)
    dst = np.float32([[offset,width * offset_l], [height-offset,width * offset_l], [height-offset,width * offset_r], [offset,width * offset_r]])
    #dst = np.reshape(dst, (4,1,2))
    #getPerspectiveTransform using src and dst
    M=cv2.getPerspectiveTransform(src,dst)
    Minv=cv2.getPerspectiveTransform(dst,src)
    warped = cv2.warpPerspective(img,M,(height, width))    
    return warped,M,Minv

def find_lane_pixels(binary_warped):
    # Take a histogram of the bottom half of the image
    histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
    # Create an output image to draw on and visualize the result
    out_img = np.dstack((binary_warped, binary_warped, binary_warped))*255
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint

    # HYPERPARAMETERS
    # Choose the number of sliding windows
    nwindows = 9
    # Set the width of the windows +/- margin
    margin = 100
    # Set minimum number of pixels found to recenter window
    minpix = 50

    # Set height of windows - based on nwindows above and image shape
    window_height = np.int(binary_warped.shape[0]//nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated later for each window in nwindows
    leftx_current = leftx_base
    rightx_current = rightx_base

    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []

    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current + margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current + margin
        # Draw the windows on the visualization image
        #cv2.rectangle(out_img,(win_xleft_low,win_y_low),(win_xleft_high,win_y_high),(0,255,0), 2) 
        #cv2.rectangle(out_img,(win_xright_low,win_y_low),(win_xright_high,win_y_high),(0,255,0), 2) 
        
        # Identify the nonzero pixels in x and y within the window #
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
        (nonzerox >= win_xleft_low) &  (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & 
        (nonzerox >= win_xright_low) &  (nonzerox < win_xright_high)).nonzero()[0]
        
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        
        # If you found > minpix pixels, recenter next window on their mean position
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))

    # Concatenate the arrays of indices (previously was a list of lists of pixels)
    try:
        left_lane_inds = np.concatenate(left_lane_inds)
        right_lane_inds = np.concatenate(right_lane_inds)
    except ValueError:
        # Avoids an error if the above is not implemented fully
        pass

    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]

    return leftx, lefty, rightx, righty, out_img


def fit_polynomial(binary_warped):
    # Find our lane pixels first
    leftx = lefty = rightx = righty = 0
    leftx, lefty, rightx, righty, out_img = find_lane_pixels(binary_warped)
    width=binary_warped[1]
    height=binary_warped[0]
    left_fitx = 0
    right_fitx = 0
    ploty = 0
    # Fit a second order polynomial to each using `np.polyfit`
    left_fit = np.polyfit(lefty, leftx, 2)
    right_fit = np.polyfit(righty, rightx, 2)

    # Generate x and y values for plotting
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    try:
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    except TypeError:
        # Avoids an error if `left` and `right_fit` are still none or incorrect
        print('The function failed to fit a line!')
        left_fitx = 1*ploty**2 + 1*ploty
        right_fitx = 1*ploty**2 + 1*ploty

    ## Visualization ##
    # Colors in the left and right lane regions
    out_img[lefty, leftx] = [255, 0, 0]
    out_img[righty, rightx] = [255, 0, 0]

    # Plots the left and right polynomials on the lane lines
    plt.plot(left_fitx, ploty, color='yellow', linewidth='20')
    plt.plot(right_fitx, ploty, color='yellow',linewidth='20')
    start_point = 100
    for point in range(len(left_fitx)):
        start_point = start_point + 1
        cv2.line(out_img,((left_fitx[point]).astype(np.int32), start_point),((right_fitx[point]).astype(np.int32),start_point),(0,200,0),10)

    return out_img,lefty,leftx,righty,rightx,left_fitx,right_fitx,ploty

def fillRoadLanes(out_img,leftx_low,lefty_low,leftx_high,lefty_high,rightx_high,righty_high,rightx_low,righty_low):
    cv2.fillPoly(out_img,[np.array([[leftx_low,lefty_low],[leftx_high,lefty_high],[rightx_high,righty_high],[rightx_low,righty_low]])],(0,200,0))
    return out_img
    
def inversePerspective(outimg, Minv):
    cv2.warpPerspective(outimg,Minv,(outimg.shape[1],outimg.shape[0]))
    return out_img

def weightedImage(undistortedimage, unwarpedoverlay):
    return cv2.addWeighted(undistortedimage,1,unwarpedoverlay,3,0)

def calculateDistance(img,left_fit,right_fit):
    '''Assumption is that the lane length is 3.7meters wide and camera is fitted at the center of vehicle'''
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    # Assuming camera is centered in the image, return position of car in cms.
    car_loc = img.shape[1] / 2

    # Find lane line bottom x-corrdinate
    height = img.shape[0]
    centre_offset = ((right_fit[-1] - car_loc)-(car_loc - left_fit[-1]))*xm_per_pix
    return centre_offset

def calculateCurvature(ploty,left_fit_x,right_fit_x):
    # Define conversions in x and y from pixels space to meters
    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    # Define y-value where we want radius of curvature
    # We'll choose the maximum y-value, corresponding to the bottom of the image
    y_eval = np.max(ploty)
    left_fit_cr = np.polyfit(ploty * ym_per_pix, left_fit_x*xm_per_pix,2)
    right_fit_cr = np.polyfit(ploty * ym_per_pix, right_fit_x*xm_per_pix,2)
    ##### TO-DO: Implement the calculation of R_curve (radius of curvature) #####
    left_curverad = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    right_curverad = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    return left_curverad, right_curverad
```

```python
mtx, dist = calibrateCamera(9,6)
st = 0
def process_image(image):
    #st=st+1
    undistimg = undistortImage(image,mtx,dist)
    plt.imshow(undistimg)
    #Gaussian blur
    blurredimage = gaussianBlur(undistimg, 5)
    #combined binary image
    colorbinary, combinedbinary = combinedBinary(blurredimage,50,150,100,255)
    binaryimg, pers_mtx, inv_mtx = unwrapImage(combinedbinary)
    out_img,lefty,leftx,righty,rightx,leftfitx,rightfitx,ploty = fit_polynomial(binaryimg)
    #out_img=fillRoadLanes(out_img,leftx[0],lefty[0],leftx[-1],lefty[-1],rightx[-1],righty[-1],rightx[0],righty[0])
    unwarped_overlay = cv2.warpPerspective(out_img,inv_mtx,(out_img.shape[1],out_img.shape[0]))
    finalimg = cv2.addWeighted(undistimg,1,unwarped_overlay,3,0)
    leftcurv,rightcurv=calculateCurvature(ploty,leftfitx,rightfitx)
    distancefromcentre=calculateDistance(finalimg,leftx,rightx)
    # describe the type of font 
    # to be used. 
    font = cv2.FONT_HERSHEY_SIMPLEX 
  
    # Use putText() method for 
    # inserting text on video 
    cv2.putText(finalimg,'Left Curvature:'+str(round(leftcurv,2))+' Right Curvature:'+str(round(rightcurv,2)),(50, 50),font, 1,(255, 255, 255),  
                2,  
                cv2.LINE_4) 
    cv2.putText(finalimg,'Distance:'+str(round(distancefromcentre,2)),(50, 150),font, 1,(255, 255, 255),  
                2,  
                cv2.LINE_4) 
    #cv2.imwrite('pout'+str(st)+'.jpg',finalimg)
    
    #mpimg.imwrite("projectim.jpg")
    return finalimg
```
